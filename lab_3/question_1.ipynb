```python
import tensorflow as tf
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
from keras.models import Sequential
import keras.layers
```

```python
X,y = make_circles(n_samples=1000, factor=0.5, noise = 0.1)
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
nn = Sequential()
nn.add(keras.layers.Dense(units=5, activation='relu', input_shape=(2,)))
nn.add(keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
nn.compile(optimizer='sgd', loss = keras.losses.BinaryCrossentropy(), metrics=keras.metrics.BinaryAccuracy())
```

```python
nn.fit(X, y, batch_size=16, epochs=20, verbose=True)
```
