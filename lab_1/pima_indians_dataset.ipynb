```python
#Yadukrishnan - 21BAI1210

import pandas as pd
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
#Loading csv file using pandas
diabetes_data = pd.read_csv('/kaggle/input/pima-indians-diabetes-database/diabetes.csv')

#Splitting the columns and target columns
X = pd.DataFrame(diabetes_data.drop(["Outcome"],axis = 1),)
y = diabetes_data.Outcome

#Splitting train and test dataset
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=1/3)
```

```python
#Defining PCA
pca = PCA(n_components=2)

#Feeding values into PCA
pca.fit_transform(X_train)
for i, variance_ratio in enumerate(pca.explained_variance_ratio_):
    print(f"Principal Component {i + 1}: {variance_ratio * 100:.2f}%")
```

```python
#Transform train dataset with the PCA
X_train_transformed = pca.transform(X_train)
X_test_transformed = pca.transform(X_test)
```

```python
#Defining KNN for classification
knn = KNeighborsClassifier(3)

#Feeding dataset into KNN
knn.fit(X_train_transformed, y_train)

#Accuracy of train and test dataset - 21BAI1210
print("Accuracy of training data:", knn.score(X_train_transformed, y_train))
print("Accuracy of test data:", knn.score(X_test_transformed, y_test))
```
